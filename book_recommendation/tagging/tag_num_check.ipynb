```python
import json

with open("data/tagged_books.json", "r") as file:
    books = json.load(file)
```

```python
h_title = [book.get('title') for book in books if book.get('tag') == '흥미진진한']
s_title = [book.get('title') for book in books if book.get('tag') == '서늘한']
y_title = [book.get('title') for book in books if book.get('tag') == '유쾌한']
j_title = [book.get('title') for book in books if book.get('tag') == '잔잔한']

print("흥미진진한 개수: ", len(h_title), "제목:", h_title)
print("서늘한 개수: ", len(s_title), "제목:", s_title)
print("유쾌한 개수: ", len(y_title), "제목:", y_title)
print("잔잔한 개수: ", len(j_title), "제목:", j_title)
```
